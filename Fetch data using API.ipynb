```python
# import time
import requests

new_list = []
for j in range(1, 114):
    try:
        url = f'https://api.themoviedb.org/3/tv/top_rated?api_key=d0b8333ad911dd05f71c4eddecd0f830&language=en-US&page={j}'
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # raises error for bad HTTP status codes

        data = response.json()['results']
        for item in data:
            temp_dict = {
                'id': item['id'],
                'Title': item['name'],
                'Popularity': item['popularity'],
                'Vote_count': item['vote_count'],
                'Original_language': item['original_language'],
                'Overview': item['overview']
            }
            new_list.append(temp_dict)

        # time.sleep(0.3)  # wait 300ms before next request

    except requests.exceptions.RequestException as e:
        # print(f"Error on page {j}: {e}") 
        continue
```

```python
movies_df = pd.DataFrame(new_list)
```
