### Pandas Export function.

- to_csv()
- to_excel()

#### Import Required Libraries.

```python
import pandas as pd
```

```python
deliveries = pd.read_csv('./Datasets/deliveries.csv')
```

```python
deliveries.head()
```

### Export using to_csv() function.

#### `Question` Fetch all the batsman runs scored in IPL.

```python
temp_df = deliveries.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_csv('./Export_Datasets/batsman_runs.csv', index=False)
```

#### `Question` Find top ten batsmans who score highest runs in IPL.

```python
temp_df = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(10)
temp_df
```

```python
temp_df.to_csv('./Export_Datasets/Top_ten_batsman.csv', index=False)
```

### Export using to_excel function.

```python
zomato = pd.read_csv('./Datasets/zomato.csv', encoding='latin-1')
```

```python
zomato.head()
```

#### `Question` Find all the order that are done online.

```python
temp_df1 = zomato[zomato['Has Online delivery'] == 'Yes']
temp_df1
```

#### `Question` Find all the order that are done offline.

```python
temp_df2 = zomato[zomato['Has Online delivery'] == 'No']
temp_df2
```

```python
with pd.ExcelWriter('./Export_Datasets/Zomato_booking.xlsx') as writer: 
    temp_df1.to_excel(writer, sheet_name='Online Orders')
    temp_df2.to_excel(writer, sheet_name='Offline Orders')
```

### Export using to_html() function

```python
temp_pivot_table = deliveries.query("batsman_runs == 6").pivot_table(index='over', columns='ball', values='batsman_runs', aggfunc='count')
temp_pivot_table
```

```python
temp_pivot_table.to_html('./Export_Datasets/sixes_pivot_table.html')
```

### Export using to_json function

```python
temp_df = deliveries.groupby(['batting_team', 'batsman'])['batsman_runs'].sum().unstack()
```

```python
temp_df.to_json('./Export_Datasets/Batsman_runs_for_each_team.json')
```

### Export dataframe using to_sql() function.

```python
import pymysql
from sqlalchemy import create_engine
```

```python
# Syntax for this argument in create_engine("mysql+pymysql://root:{password}@{url}/{database_name}")
engine = create_engine("mysql+pymysql://root:@localhost/ipl")
deliveries.to_sql('ipl_deliveries', con=engine, if_exists='append')
```

```python
temp_df = deliveries.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df.to_sql('batsman_runs', con=engine, if_exists='append', index=False)
```
